```python
import sys
sys.path.append('../')

from lib import libdata
import os
import numpy as np
import pandas as pd
```

```python
transcirc_circrnas_file = 'TransCirc/transcirc_sequence.fa.bgz'
transcirc_orfs_file = 'TransCirc/transcirc_orf_na.fa.bgz'

circrnas_fasta_file = libdata.CIRCODAN_PATH + 'datasets/03_TransCirc_RP_PP_evidence_hsa.fa'
corfs_gtf_file = libdata.CIRCODAN_PATH + 'datasets/03_TransCirc_RP_PP_evidence_hsa.gtf'

output_path = libdata.CIRCODAN_PATH + 'datasets/TransCirc_RP_PP_splitted/'
```

```python
df_circs = libdata.read_fasta_in_dataframe(circrnas_fasta_file)

# Shuffle the circRNAs
df_circs_shuffled = df_circs.sample(frac=1, random_state=721379)

# Split the circRNAs into 5 equal parts
splits = np.array_split(df_circs_shuffled, 5)

splits[0]
```

```python
fasta_file_name = 'TransCirc_RP_PP_evidence_hsa.fa'
gtf_file_name = 'TransCirc_RP_PP_evidence_hsa.gtf'

for i in range(5):

    print('\nSplit:\t', i)

    test_ids = []
    train_ids = []

    for j in range(5):

        if j == i:
            test_ids = list(splits[j]['seqname'])
        else:
            train_ids += list(splits[j]['seqname'])

    print ('test:\t', len(test_ids))
    print ('train:\t',len(train_ids))

    split_path = output_path + str(i) + '/'
    os.mkdir(split_path)

    test_file_path = split_path + 'test_' + fasta_file_name
    train_file_path = split_path + 'train_' + fasta_file_name

    # Test
    libdata.create_fasta_file_TransCirc(test_ids, transcirc_circrnas_file, test_file_path)
    print('circRNAs for test:', libdata.count_sequences_in_file(test_file_path))
    libdata.create_gtf_file_TransCirc(test_file_path, transcirc_orfs_file, test_file_path.replace('.fa', '.gtf'))
    df_gtf = pd.read_csv(test_file_path.replace('.fa', '.gtf'), sep='\t', header=None)
    df_gtf.columns = libdata.get_gtf_columns()
    print('cORFs for test:', df_gtf.shape)

    # Train
    libdata.create_fasta_file_TransCirc(train_ids, transcirc_circrnas_file, train_file_path)
    print('circRNAs for train:', libdata.count_sequences_in_file(train_file_path))
    libdata.create_gtf_file_TransCirc(train_file_path, transcirc_orfs_file, train_file_path.replace('.fa', '.gtf'))
    df_gtf = pd.read_csv(train_file_path.replace('.fa', '.gtf'), sep='\t', header=None)
    df_gtf.columns = libdata.get_gtf_columns()
    print('cORFs for train:', df_gtf.shape)
```
