```python
import sys
sys.path.append('../')

from lib import libdata
import pandas as pd
```

```python
# Input file
raw_data_file = 'MStoCIRC/result.mstocirc.hsa/circ_draw.txt'

# Output files
circrnas_file = libdata.CIRCODAN_PATH + 'datasets/06_MStoCIRC_hsa.fa'
gtf_file = libdata.CIRCODAN_PATH + 'datasets/06_MStoCIRC_hsa.gtf'
```

```python
cols = ['circRNAname', 'circRNA_length', 'CORF_start', 'CORF_end', 'IRES_start', 'IRES_end', 'exon_position', 'informa', 'nt_sequence', 'aa_sequence']
df_data = pd.read_csv(raw_data_file, header=None, names=cols, comment='#')
df_data
```

```python
libdata.create_fasta_file_MStoCIRC(df_data['circRNAname'], df_data['nt_sequence'], circrnas_file)
libdata.count_sequences_in_file(circrnas_file)
```

```python
# Removing repeat sequences
n_unique_seqs = libdata.remove_repeated_sequences_in_file(circrnas_file)
n_unique_seqs
```

```python
# Getting cORF end and translation_cycles
orf_end_positions = []
for _,row in df_data.iterrows():
    orf_end_positions.append(libdata.get_corf_end_position_zi(row['CORF_start']-1, row['nt_sequence']) + 1)
df_data['CORF_end_corrected'] = orf_end_positions

df_data['translation_cycles'] = df_data['CORF_end'] // df_data['circRNA_length']

df_data
```

```python
# Checking start codon
df_data['start_codon'] = df_data.apply(lambda row: row['nt_sequence'][row['CORF_start']-1:row['CORF_start']+2], axis=1)
df_data['start_codon_correct'] = df_data['start_codon'].apply(lambda sequence: sequence[-3:] in ['ATG'])
df_data[~df_data['start_codon_correct']]
```

```python
# Fixing start codon
df_data.loc[df_data['circRNAname'] == 'hsa_circ_0010401', ['CORF_start', 'CORF_end_corrected']] = [1, 22]
df_data.loc[df_data['circRNAname'] == 'hsa_circ_0010401']
```

```python
libdata.create_gtf_file_MStoCIRC(df_data, gtf_file)
```
