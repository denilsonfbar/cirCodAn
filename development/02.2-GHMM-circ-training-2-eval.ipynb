```python
import pandas as pd

results_file_path = 'results/02.tsv'
```

```python
df_results = pd.read_csv(results_file_path, sep='\t', header=0)
df_results[['tool', 'model', 'prob', 'split', 'input_file', 'F1-s_full_seq']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_plot = df_results
column_names = {'F1-s_full_seq':'F1-score', 'Prec_full_seq':'Precision', 'Rec_full_seq':'Recall'}
df_plot.rename(columns=column_names, inplace=True)

x_axis_col = 'prob'                        
y_axis_cols = ['F1-score', 'Precision', 'Recall']

dims = (8, 4.5)
plt.subplots(figsize=dims)
plt.grid()

for col in y_axis_cols:
    sns.lineplot(data=df_plot, x=x_axis_col, y=col, legend='brief', label=col, marker='o')
    plt.xscale('linear')

y_values = list(range(0, 101, 10))
plt.yticks(y_values)

plt.ylabel('Percentage of evaluation metrics')
plt.xlabel('Transition probability between NOCOD and START states')
```
